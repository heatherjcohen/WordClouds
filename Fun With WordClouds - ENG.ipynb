```python
#### Emperor's New Groover in the shape of a llama 

import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import numpy as np
url = 'http://www.script-o-rama.com/movie_scripts/e/emperors-new-groove-script-transcript.html'
response = requests.get(url)

```

```python
soup = BeautifulSoup(response.text, "html.parser")
words = soup.pre.get_text()
```

```python
words = soup.pre.get_text()
words= re.sub("[^a-zA-Z' ]+", " ", words)
```

```python
#identify custom stop words
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

value_counts= word_count(words)

for key, value in sorted(value_counts.items(), reverse=True, key=lambda item: item[1]):
    print("%s: %s" % (key, value))
```
