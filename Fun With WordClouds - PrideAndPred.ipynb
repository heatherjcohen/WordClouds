```python
#P&P
from os import path
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from PIL import ImageEnhance
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import numpy as np
```

```python
stopwords = set(STOPWORDS)
#stopwords.add('dah')

  
```

```python
def cloudify (url, maskfile):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    words = soup.get_text()
    words= re.sub("[^a-zA-Z' ]+", " ", words)
    mask = np.array(Image.open(maskfile))
    wordcloud = WordCloud(stopwords = stopwords, background_color="white", max_words=3000, mask=mask).generate(words)
    image_colors = ImageColorGenerator(mask)
    plt.figure(figsize=[8,8])
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.axis("off")
    #plt.savefig("stevesie.png", format="png")
    plt.show()
    plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
    plt.axis("off")
    plt.show()


    
```

```python
cloudify('http://www.gutenberg.org/files/1342/1342.txt', "jane-austen_in_blue_dress_e5nojpg.jpg")
```

```python
def getcorpus(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    words = soup.get_text()
    words= re.sub("[^a-zA-Z' ]+", " ", words)
    return words
    
```

```python
words = getcorpus('http://www.gutenberg.org/files/1342/1342.txt')
```

```python
def cloud(text,maskfile):
    mask = np.array(Image.open(maskfile))
    wordcloud = WordCloud(stopwords = stopwords, background_color="white", max_words=3000, mask=mask).generate(text)
    image_colors = ImageColorGenerator(mask)
    plt.figure(figsize=[8,8])
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.axis("off")
    plt.savefig("viz"+str(maskfile), format="png")
    plt.show()
    plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
def cloudtweek(text,maskfile,maxsize):
    mask = np.array(Image.open(maskfile))
    wordcloud = WordCloud(stopwords = stopwords, background_color="white", max_words=3000,max_font_size=maxsize,  mask=mask).generate(text)
    image_colors = ImageColorGenerator(mask)
    plt.figure(figsize=[8,8])
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.axis("off")
    plt.savefig("viz"+str(maskfile), format="png")
    plt.show()
    plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
cloudtweek(words, "jane-austen_in_blue_dress_e5nojpg.jpg",100)
```
