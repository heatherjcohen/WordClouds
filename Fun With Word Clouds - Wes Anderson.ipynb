```python

    
#Wes Anderson poster wordclouds  
from os import path
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from PIL import ImageEnhance
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import numpy as np
url = 'http://www.script-o-rama.com/movie_scripts/r/royal-tenenbaums-script-transcript-anderson.html'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
words = soup.pre.get_text()
```

```python
words = soup.pre.get_text()
words= re.sub("[^a-zA-Z' ]+", " ", words)
```

```python
#Make a square poster based on the colors


stopwords = set(STOPWORDS)
stopwords.add('dah')
stopwords.add('know')
stopwords.add('think')
stopwords.add('yes')
stopwords.add('want')
stopwords.add('say')
stopwords.add('Okay')
stopwords.add('Oh')
stopwords.add('Yeah')
stopwords.add('Uh')
stopwords.add('right')
stopwords.add('well')
stopwords.add('got')
stopwords.add('going')



mask = np.array(Image.open("e37fe0953128bff211ef10c7c064e656.jpg"))
wordcloud_por = WordCloud(stopwords = stopwords, background_color="white", max_words=3000, mask=mask).generate(words)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wordcloud_por.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.figure()
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
